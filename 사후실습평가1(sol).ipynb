# Q2. min-max scailing 후, 12번째 row의 s2컬럼의 값은 무엇인가?

**조건**
1. S1~ S6 만 스케일링 할 것
1. 소수점 세자리까지, 그 미만 절삭.
2. 0번째 row부터 있습니다.

```
# x.head()
# x.columns
# list(x)
list(x)[-6:]
```

```
list(x)[-6:]
```

```
for col in list(x)[-6:] :
    min_x = x[col].min()
    max_x = x[col].max()
    boonmo = max_x - min_x
    x[col] = (x[col]-min_x)/boonmo
    
x.describe()
```

```
print('s2 : ', x.loc[12,'S2'])
print('답은 0.336')
```
