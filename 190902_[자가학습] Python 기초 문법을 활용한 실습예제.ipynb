## 6. Slicing
- slicing list
- slicing string

```
# 리스트 생성
numbers = [95, 65, 90, 80, 85, 90, 92, 88, 84, 75, 85, 100]

print('')
print(numbers)
print('---------')

# 리스트의 subset 01
subset = numbers[:5]
print(subset)

# 리스트의 subset 02
subset = numbers[5:10]
print(subset)

# 리스트의 subset 03
subset = numbers[-5:]
print(subset)
```

```
# 문자열 생성
python = "hobby project during the week around Christmas as his office was closed"
    
print('')
print(python)
print('---------')

# 문자열 subset 01
print(python[:5])

# 문자열 subset 02
print(python[14:20])

# 문자열 subset 03
print(python[-6:])
```

## 2. Control (제어문)

- 조건문: if-else
- 반복문: while
- 반복문: for


```
메뉴1 = ["진주물냉면", 8500]
메뉴2 = ["진주비빔냉면", 8500]
메뉴3 = ["남해멸치쌈밥", 10000]
메뉴4 = ["만두", 5000]
메뉴목록 = [메뉴1, 메뉴2, 메뉴3, 메뉴4]

주문1 = []
주문1.append(메뉴목록[0])
주문1.append(메뉴목록[0])
주문1.append(메뉴목록[1])

주문2 = []
주문2.append(메뉴목록[2])
주문2.append(메뉴목록[2])

# 조건문 - if/else
print(주문1, 주문2)
가진돈 = 20000

# 주문액이 가진돈 보다 작으면 '주문할 수 있다.'
# 주문액이 가진돈 보다 크면 '돈이 모자란다.'
주문1_합계 = 주문1[0][1] + 주문1[1][1] + 주문1[2][1] 
if 주문1_합계 <= 가진돈:
    print('주문할 수 있다.')
else:
    print('돈이 모자란다')
    
주문2_합계 = 주문2[0][1] + 주문2[1][1]
if 주문2_합계 <= 가진돈:
    print('주문할 수 있다.')
else:
    print('돈이 모자란다')
```

```
# 반복문 - for
print(주문1, 주문2)

# 주문합계 계산
주문1_합계 = 0
for 주문 in 주문1:
    print(주문)
    주문1_합계 += 주문[1]

주문2_합계 = 0    
print(주문1_합계, 주문2_합계)

'''
[['진주물냉면', 8500], 
 ['진주물냉면', 8500], 
 ['진주비빔냉면', 8500]]
'''
```

### 과제
- 로또 번호 생성기를 만들어보자
 - 1 ~ 46까지의 수 중 6개의 숫자를 뽑는다.
 - 숫자는 중복될 수 없다.
- 가위바위보 게임을 만들어보자 
 - 컴퓨터와 가위바위보를 한다.
 - 승부는 3판 2선 승!


```
import random

# 로또 번호를 담을 변수에 빈 리스트를 만들어요.
lotto = []

# 리스트가 6개가 될때까지 반복하자.
while len(lotto) < 6:
    # 1~46 중에 랜덤한 숫자 하나를 선택
    num = random.randint(1, 46)

    # 리스트에 뽑은 숫자가 없으면, 리스트에 넣는다
    if num not in lotto:
        lotto.append(num)
        
print(lotto)
```
