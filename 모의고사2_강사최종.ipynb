```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
x = boston.data
y = boston.target

x.shape, y.shape
```

```python
y = y.reshape( [ -1 , 1])
```

```python
data = pd.DataFrame(data = x,
                   columns = boston.feature_names)

data.head()
```

```python
cols = list(data)
```

```python
cols.remove(    'CHAS')
# try except, catch
```

```python
for each in cols :
    min_each = data[each].min()
    max_each = data[each].max()
    boonmo = max_each - min_each
    data[each] = (data[each]-min_each)/boonmo
    
data.describe()
```

```python
x = data_origin.values
```

# Q3. 모델링을 하자!

* Linear Regression을 한다!

* KNN을 이용한 Regression 도 해본다!

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
lr = LinearRegression()
kr = KNeighborsRegressor()
```

```python
lr.fit(x, y)
kr.fit(x, y)
```

```python
l_pred = lr.predict(x)
k_pred = kr.predict(x)
```

```python
print(l_pred)
print(k_pred)
```

# Q4. 평가하자!
* MSE 혹은 RMSE를 구해보자!
* R^2 값을 구해보자! ( Scikit-learn 의 r2_score 참고!)

### 쉬운길

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
print("선형회귀의 rmse : {:.3f}".format(np.sqrt(mean_squared_error(y, l_pred))))
print("KNN회귀의 rmse : {:.3f}".format(np.sqrt(mean_squared_error(y, k_pred))))
```

```python
print("선형회귀의 rmse : {:.3f}".format(r2_score(y, l_pred)))
print("KNN회귀의 rmse : {:.3f}".format(r2_score(y, k_pred)))
```
